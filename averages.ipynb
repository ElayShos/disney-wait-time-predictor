```python
import pandas as pd
import glob
import os

def average_hourly_wait_time(park):
    path = f'data/*/wait{park.upper()}.csv'
    path = os.path.join('data','*', f'wait{park.upper()}.csv')
    files = glob.glob(path)
    df = pd.concat(pd.read_csv(f) for f in files)
    df.iloc[:, 0] = pd.to_datetime(df.iloc[:, 0])
    df.set_index(df.iloc[:, 0], inplace=True)
    df = df.iloc[:, 1:]

    morning = df.between_time('06:00', '10:00')
    late_morning = df.between_time('10:01', '12:00')
    noon = df.between_time('12:01', '14:00')
    afternoon = df.between_time('14:01', '17:00')
    evening = df.between_time('17:01', '19:00')
    night = df.between_time('19:01', '23:59')

    row = {}

    def fix_num(x):
        x = int(x) if pd.notna(x) else 0
        x = round(x/5) * 5
        return x

    for ride in df.columns:
        mo = morning[ride].replace(0, pd.NA).mean()
        lm = late_morning[ride].replace(0, pd.NA).mean()
        no = noon[ride].replace(0, pd.NA).mean()
        af = afternoon[ride].replace(0, pd.NA).mean()
        ev = evening[ride].replace(0, pd.NA).mean()
        ni = night[ride].replace(0, pd.NA).mean()

        
        mo = fix_num(mo)
        lm = fix_num(lm)
        no = fix_num(no)
        af = fix_num(af)
        ev = fix_num(ev)
        ni = fix_num(ni)

        avg_lines = [mo, lm, no, af, ev, ni]
        row[ride] = avg_lines
    new_df = pd.DataFrame.from_dict(row, orient='index', columns=['Morning', 'Late Morning', 'Noon', 'Afternoon', 'Evening', 'Night'])
    print("finished calculating average wait times")

    os.makedirs(os.path.join('analysis_results', 'average_wait_times'), exist_ok=True)
    path_save = os.path.join('analysis_results', 'average_wait_times', f'average_wait_{park.upper()}.csv')
        
    new_df.to_csv(path_save)


average_hourly_wait_time('mk')

    
```
